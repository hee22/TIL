```python
import sys
import math

# Auto-generated code below aims at helping you parse
# the standard input according to the problem statement.
def get_rank(card):
    number = card[:-1]
    if number == "J":
        return 11
    elif number == "Q":
        return 12
    elif number == "K":
        return 13
    elif number == "A":
        return 14
    else:
        return int(number)

def fight(card1, card2):
    rank1 = get_rank(card1)
    rank2 = get_rank(card2)
    if rank1 > rank2:
        return 1
    if rank1 < rank2:
        return 2
    else:
        return 0
        
def war(deck1, deck2, card1, card2):
    result = 0
    
    war_deck1 = [card1]
    war_deck2 = [card2]
    while len(deck1) > 3 and len(deck2) > 3:

        war_deck1.append(deck1.pop(0))
        war_deck1.append(deck1.pop(0))
        war_deck1.append(deck1.pop(0))
            
        war_deck2.append(deck2.pop(0))
        war_deck2.append(deck2.pop(0))
        war_deck2.append(deck2.pop(0))
        
        card1 = deck1.pop(0)
        card2 = deck2.pop(0)
        war_deck1.append(card1)
        war_deck2.append(card2)
        
        result = fight(card1, card2)
        if result == 1:
            deck1.extend(war_deck1)
            deck1.extend(war_deck2)
            break
        
        if result == 2:
            deck2.extend(war_deck1)
            deck2.extend(war_deck2)
            break
            
    
    return (result, deck1, deck2)

deck1 = []
deck2 = []
turn = 0
result = 0

n = int(input())  # the number of cards for player 1
for i in range(n):
    cardp_1 = input()  # the n cards of player 1
    deck1.append(cardp_1)
m = int(input())  # the number of cards for player 2
for i in range(m):
    cardp_2 = input()  # the m cards of player 2
    deck2.append(cardp_2)
    
while len(deck1) > 0 and len(deck2) > 0:
    turn += 1
    
    card1 = deck1.pop(0)
    card2 = deck2.pop(0)
    
    result = fight(card1, card2)
    
    if result == 1:
        deck1.extend([card1, card2])
    elif result == 2:
        deck2.extend([card1,card2])
    else:
        result, deck1, deck2 = war(deck1, deck2, card1, card2)
        if result == 0:
            break
    
if result == 0:
    print("PAT")
    
elif len(deck1) > 0:
    print(1, turn)

else:
    print(2, turn)
```
