```python
import sys
import math

# Auto-generated code below aims at helping you parse
# the standard input according to the problem statement.

message = input()

# Write an action using print
def to_binary(demical):
    binary = ""
    
    while demical > 0:
        reminder = demical % 2
        binary += str(reminder)
        demical = demical // 2
    
    if len(binary) == 6:
        binary += "0"
    
    return binary[::-1]
    
def chuck_norris_encoding(code):
    
    pre_bit = ""
    count = 0
    answer = ""
    
    for bit in code:
        if bit != pre_bit and count > 0:
            if pre_bit == "1":
                answer += "0 "
            else:
                answer += "00 "

            answer += "0" * count + " "
            count = 0
        count += 1
        pre_bit = bit
    
    if pre_bit == "1":
        answer += "0 "
    else:
        answer += "00 "
        
    answer += "0" * count
    
    return answer
    
# To debug: print("Debug messages...", file=sys.stderr)
binary_string = ""

for ch in message:
    binary_string += to_binary(ord(ch))
    
encoded_message = chuck_norris_encoding(binary_string)
print(encoded_message)
```
